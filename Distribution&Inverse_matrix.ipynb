```python
import numpy as np
import random
import matplotlib.pyplot as plt

# Generating the original population data
x = np.random.normal(loc=300.0, size=1000)

# Plotting the original population data distribution
plt.figure(figsize=(6, 4))
plt.subplot(2, 3, 1)
plt.hist(x, bins=10, color='skyblue')
plt.title("Population Data Distribution")


sample_sizes = [2, 4, 6, 8, 10]
sample_means = []

# Looping through different sample sizes
for idx, sample_size in enumerate(sample_sizes):
    sample_mean = []
    for i in range(50):
        y = random.sample(x.tolist(), sample_size)
        avg = np.mean(y)
        sample_mean.append(avg)

    sample_means.append(sample_mean)

    # Plotting the sampling distribution for each sample size
    plt.subplot(2, 3, idx + 2)
    plt.hist(sample_mean, bins=10, color='lightcoral')
    plt.title(f"Sample Size: {sample_size}")

plt.tight_layout()
plt.show()
```

```python
def determinant_2x2(matrix):
    return matrix[0][0] * matrix[1][1] - matrix[0][1] * matrix[1][0]
```

```python
def matrix_minor(matrix, row, col):
    return [[matrix[i][j] for j in range(len(matrix)) if j != col] for i in range(len(matrix)) if i != row]
```

```python
def matrix_determinant(matrix):
    if len(matrix) == 2:
        return determinant_2x2(matrix)

    determinant = 0
    for j in range(len(matrix)):
        determinant += ((-1) ** j) * matrix[0][j] * determinant_2x2(matrix_minor(matrix, 0, j))
    return determinant
```

```python
def matrix_of_cofactors(matrix):
    cofactors = []
    for i in range(len(matrix)):
        cofactor_row = []
        for j in range(len(matrix)):
            minor = matrix_minor(matrix, i, j)
            sign = (-1) ** (i + j)
            cofactor_row.append(sign * matrix_determinant(minor))
        cofactors.append(cofactor_row)
    return cofactors
```

```python
def transpose(matrix):
    return [[matrix[j][i] for j in range(len(matrix))] for i in range(len(matrix))]
```

```python
def matrix_inverse(matrix):
    determinant = matrix_determinant(matrix)
    if determinant == 0:
        raise ValueError("Matrix is not invertible (determinant is zero).")

    cofactors = matrix_of_cofactors(matrix)
    adjugate = transpose(cofactors)

    inverse = [[adjugate[i][j] / determinant for j in range(len(matrix))] for i in range(len(matrix))]

    return inverse
matrix = [[1, 2, 0], [4, 2, 5], [1, 0, 2]]
try:
    inverse_matrix = matrix_inverse(matrix)
    print("Inverse matrix:")
    for row in inverse_matrix:
        print(row)
except ValueError as e:
    print(e)
```
